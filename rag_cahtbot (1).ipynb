```
# Imports
from langchain.vectorstores import FAISS
# from langchain.embeddings import HuggingFaceEmbeddings # This line is moved to cell YhnI8bp-7jDy
from langchain.llms import HuggingFacePipeline
from langchain.chains import RetrievalQA
from transformers import pipeline
```

```
# STEP 1: Small dataset
data = [
    {"text": "Machine learning is a method where computers learn from data."},
    {"text": "Deep learning is a subset of machine learning using neural networks."},
    {"text": "Python is a popular programming language for AI and data science."},
    {"text": "FAISS is a library for efficient similarity search and clustering."},
    {"text": "LangChain helps build applications using large language models."},
    {"text": "RAG stands for Retrieval-Augmented Generation."},
    {"text": "Embeddings convert text into numerical vectors."},
    {"text": "Neural networks are inspired by the human brain."},
    {"text": "Supervised learning uses labeled data for training."},
    {"text": "Unsupervised learning finds patterns in unlabeled data."}
]

# Convert to list of texts
texts = [item["text"] for item in data]

# STEP 3: Create FAISS vector store
db = FAISS.from_texts(texts, embeddings)
```

```
# STEP 2: Create embeddings
# Uninstall potentially conflicting versions
!pip uninstall -y sentence-transformers huggingface-hub

# Install latest sentence-transformers. This should pull a compatible huggingface_hub.
!pip install sentence-transformers

# Ensure huggingface-hub is compatible with other libraries like transformers if necessary.
# If sentence-transformers already installed a recent huggingface-hub, this will do nothing or upgrade slightly.
!pip install --upgrade huggingface-hub

# Re-import HuggingFaceEmbeddings after installations to ensure it's defined
from langchain.embeddings import HuggingFaceEmbeddings
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```
# STEP 4: Initialize the LLM (Large Language Model)
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline

# Choose a small, efficient model for demonstration purposes
# Note: Larger, more powerful models would require more resources (GPU RAM, longer download)
model_name = "distilbert/distilgpt2" # A small, fast GPT-2 model

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# Create a text generation pipeline
pipeline = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=50 # Limit output length for quicker responses
)

# Initialize LangChain's HuggingFacePipeline
llm = HuggingFacePipeline(pipeline=pipeline)
```

```
# STEP 4: Initialize the LLM (Large Language Model)
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline

# Choose a small, efficient model for demonstration purposes
# Note: Larger, more powerful models would require more resources (GPU RAM, longer download)
model_name = "distilbert/distilgpt2" # A small, fast GPT-2 model

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# Create a text generation pipeline
pipeline = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=50 # Limit output length for quicker responses
)

# Initialize LangChain's HuggingFacePipeline
llm = HuggingFacePipeline(pipeline=pipeline)
```

```
# STEP 5: Create a Retrieval-Augmented Generation (RAG) chain
rqa = RetrievalQA.from_chain_type(llm=llm, chain_type="stuff", retriever=db.as_retriever())

# Test the RAG chain
query = "What is machine learning?"
response = rqa.invoke(query)
print(response)

query = "What is FAISS?"
response = rqa.invoke(query)
print(response)

query = "What are neural networks?"
response = rqa.invoke(query)
print(response)
```
